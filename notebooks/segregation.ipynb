```python
cov = np.array([[1, 0], [0, 1]])

pts = np.random.multivariate_normal([0,0], cov, size=800)

pts = np.trunc(pts*100)

plt.plot(pts[:, 0], pts[:, 1], '.', alpha=0.5)
plt.axis('equal')
plt.grid()
plt.show()  
```
